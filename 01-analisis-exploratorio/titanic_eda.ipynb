# Análisis Exploratorio de Datos — Titanic

**Objetivo:** Comprender la estructura y distribución del conjunto de datos del Titanic mediante un análisis exploratorio sistemático (EDA), identificando patrones relacionados con la supervivencia de los pasajeros.

**Dataset:** [Titanic — Machine Learning from Disaster (Seaborn built-in)](https://seaborn.pydata.org/generated/seaborn.load_dataset.html)
